```python
import numpy as np
import random
import cv2
import os
from PIL import Image
```

```python
im_path="D:/PWN/Pyton/cw/data/"

def get_2d_array(im_path):
    """Return image scaled with factor 0.5 as numpy array."""
    im_color = cv2.imread(im_path)              # Read the image as a numpy array.
    im_color = im_color[32:64 + 32, 32:64 + 32] # Shape = (64,64,3) (x_pixels,y_pixels, color_channels)
    im_color = cv2.resize(im_color, (28,28))    # Rescale the image
    im = np.zeros(shape=(28,28,1))              # Create an empty array with the final shape (28,28,1)
   
    for i, x in enumerate(im_color):  # Fill the array
        im=cv2.cvtColor(im_color, cv2.COLOR_BGR2GRAY)
    return im      

def get_label(name):
    return int(name.split("_")[1])       # Returns label number for file name >>> get_label("class_10_Index_3454.jpeg") ->10
    

def main():
    n_labels = 62
    n_images = 193242
    path = "D:/PWN/Pyton/cw/data/"                 # Path to images
    images = np.zeros(shape=(n_images, 28, 28))    # Array with all images
    labels = np.zeros(shape=(n_images, n_labels))  # Array with all labels

    # Convert images to numpy arrays
    for i, file in enumerate(os.listdir(path)):
        label = get_label(file)
        image = get_2d_array(os.path.join(path, file))
        images[i] = image
        labels[i] = label
        labels[i, label] = 1
        print("Done {:.2f}".format(i / n_images * 100)+"%")
        
    # Save arrays
    np.save("nist_labels_28x28", labels)
    np.save("nist_images_28x28", images)


if __name__ == "__main__":
    main()
```
